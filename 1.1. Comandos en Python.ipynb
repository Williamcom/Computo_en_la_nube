Complex numbers (Números complejos)

```python
c = 3.14 + 2.73j # una forma de declararlo
print(c)
c = complex(3.14, 2.73) # Lo mismo que arriba
print(c)
print(c.real) # parte real
print(c.imag) # parte imaginaria
print(c.conjugate()) # para conjugar
print(c * 2) # se permite multiplicar
print(c ** 2) # y exponenciales
d = 1 + 1j # ademas de operaciones de suma y resta
print(c - d)
```

Fractions and decimals (Fracciones y decimales)

```python
from fractions import Fraction # Librería para funciones
print(Fraction(10, 6)) # declaramos una fracción
print(Fraction(1, 3) + Fraction(2, 3)) # sumamos
f1 = Fraction(10, 6) # Guardamos un fraccion
print(f1) # lo imprimimos
print(f1.numerator) # para conocer el numerador y el divisor
print(f1.denominator)
print(f1.as_integer_ratio())
```

String formatting (Formando Sting)

```python
greet_old = 'Hola %s!' # la base
greet_old %= 'Fabrizio' # una forma de cambir el dato
print(greet_old)
greet_positional = 'Hola {}!'
print(greet_positional.format('Fabrizio')) # con format en la impresion cambia
greet_positional = 'Hola {} {}!'.format('Fabrizio', 'Romano') # o al momento de declararse
print(greet_positional)
greet_positional_idx = 'esto es {0}! {1} le gusta {0}!' # tambien se puede especificar cual desea
print(greet_positional_idx.format('Python', 'Heinrich'))
print(greet_positional_idx.format('cafe', 'Fab'))
keyword = 'Hola, mi nombre es {name} {last_name}' # e incluso nombrarlo
print(keyword.format(name='Fabrizio', last_name='Romano'))
```

The global and nonlocal statements (declaraciones global y locales)

```python
def Afuera():
  test = 1 # dato de afuera

  def dentro():
    nonlocal test
    test = 2 # cambia el dato mas cercano

    print('dentro:', test)
  dentro()
  print('afuera:', test)

test = 0 # dato global
Afuera()
print('global:', test)
```

global statement (estado global)

```python
def Afuera():
  test = 1 # dato de afuera

  def dentro():
    global test
    test = 2 # cambia el dato global
    print('dentro:', test)

  dentro()
  print('Afuera:', test)

test = 0 # dato global
Afuera()
print('global:', test)
```

Assignment to parameter names (Asignando nombres de parámetros)

```python
x = 3
def func(x):
  x = 7 # este cambio local no cambiara el global
func(x)
print(x)
```

Changing a mutable object (Cambiando un objeto mutable)

```python
x = [1, 2, 3]
def func(x):
  x[1] = 42 # afecta  al original

func(x)
print(x)
```

Optional parameters (parámetros opcionales)

```python
def func(a, b=4, c=88): # al que se le asigne un valor pasa a ser un parametro opcional
  print(a, b, c)
func(1)
func(b=5, a=7, c=9)
func(42, c=9)
func(42, 43, 44)
```

Variable positional parameters (parámetros variables posicionales)

```python
def minimum(*n):
  if n: # siempre que n no este vacio
    mn = n[0]
    for value in n[1:]: #inicia la busqueda del menor
      if value < mn:
        mn = value
    print(mn)
minimum(1, 3, -7, 9)
minimum()
```
